### Tweets

```python
df = pd.read_csv('/content/drive/MyDrive/JU PROJ/Teslatweets_ALL.csv')
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.drop(['User'], axis = 1,inplace=True)
df
```

```python
df['Date'] = pd.to_datetime(df['Date']).dt.date
df
```

```python
def tweets_per_day(df):
    df['Date'] = pd.to_datetime(df['Date'])
#   return df[['Tweets']].groupby(df['Date'].dt.date).count()
    tweet = df['Tweet'].groupby(df['Date'].dt.date).count() 
    d1 = pd.DataFrame(tweet)
    return d1
  # if you want output to be `Series`

df1 = tweets_per_day(df)
df1 = df1.reset_index(level=0)
df1['fraction'] = df1['Tweet']*20000/482375
df1['fraction'] = df1['fraction'].astype(int)
df1['Date'] = pd.to_datetime(df1['Date'])
dfx = pd.DataFrame(columns=['Date', 'Tweet'])
for i in range (0,1952):

  d = df1['Date'][i]
  f = df1['fraction'][i]

  x = df[df['Date'] == d]
  x = x.sample(n = f)
  dfx = dfx.append(x, ignore_index = True)
dfx
df = dfx
df
```

```python
import nltk 
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
import re

# adding empty sentiment columns to stock_data for later calculation
df['compound'] = ''
df['negative'] = ''
df['neutral'] = ''
df['positive'] = ''

stop_words = stopwords.words('english')
stemmer = SnowballStemmer('english')

text_cleaning_re = "@\S+|https?:\S+|http?:\S|[^A-Za-z0-9]+"

def preprocess(text, stem=False):
  text = re.sub(text_cleaning_re, ' ', str(text).lower()).strip()
  tokens = []
  for token in text.split():
    if token not in stop_words:
      if stem:
        tokens.append(stemmer.stem(token))
      else:
        tokens.append(token)
  return " ".join(tokens)

df.Tweets = df.Tweet.apply(lambda x: preprocess(x))
df
```

```python
# importing requires libraries to analyze the sentiments
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import unicodedata

# instantiating the Sentiment Analyzer
sid = SentimentIntensityAnalyzer()

# calculating sentiment scores
df['compound'] = df['Tweet'].apply(lambda x: sid.polarity_scores(x)['compound'])
df['negative'] = df['Tweet'].apply(lambda x: sid.polarity_scores(x)['neg'])
df['neutral'] = df['Tweet'].apply(lambda x: sid.polarity_scores(x)['neu'])
df['positive'] = df['Tweet'].apply(lambda x: sid.polarity_scores(x)['pos'])

# displaying the stock data
df
```

```python
# Set date as Index
df = df.set_index('Date')
df = df.sort_index()
df
```
