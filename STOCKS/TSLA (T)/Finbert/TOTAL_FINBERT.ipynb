```python
import pandas as pd
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/JU PROJ/TSLA FinBERT/EXEC_FINBERT_Tesla_stock_data.csv', index_col = False)
df1
```

```python
df2 = pd.read_csv('/content/drive/MyDrive/JU PROJ/TSLA FinBERT/GENERAL_FINBERT_Tesla_stock_data.csv', index_col = False)
df2
```

```python
# Set date as Index
df1 = df1.set_index('Date')
df1 = df1.sort_index()
df2 = df2.set_index('Date')
df2 = df2.sort_index()
df1
```

```python
df2
```

```python
df3 = pd.concat((df1, df2))
df3 = df3.groupby(df3.index).mean()
df3
```
