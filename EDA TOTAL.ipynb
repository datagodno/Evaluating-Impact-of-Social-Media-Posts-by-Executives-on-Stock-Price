## Total

```python
pdList = [tsla, aapl, btc, eth]  # List of your dataframes
twitter = pd.concat(pdList)
twitter
```

```python
twitter = twitter.drop_duplicates(subset=['Tweet'], keep='first')
twitter = twitter.reset_index(drop=True)
twitter
```

```python
print(twitter['Executive/Non-executive'].value_counts())
```

```python
print(twitter['STOCK'].value_counts())
```

```python
x = twitter['STOCK'][twitter['Executive/Non-executive'] == 'Executive'].value_counts()
```

```python
y = twitter['STOCK'][twitter['Executive/Non-executive'] == 'Non-executive'].value_counts()
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(x,'g-')
ax2.plot(y, 'b-')

ax1.set_xlabel('STOCKS')
ax1.set_ylabel('Executive', color='g')
ax2.set_ylabel('Non-Executive', color='b')

plt.show()
```

```python
sns.countplot(x='STOCK', data=twitter,order=['BTC','TSLA','ETH','AAPL'])
```

```python
filter_list = ['Executive','Non-executive']
df = twitter[twitter['Executive/Non-executive'].isin(filter_list)]

fig = plt.gcf()
sns.countplot(x='STOCK', hue='Executive/Non-executive', data=df, order = df['STOCK'].value_counts().index, 
              hue_order = df['Executive/Non-executive'].value_counts().index)
```

```python
twitter['Executive/Non-executive'].value_counts().plot.pie()
plt.gca().set_aspect('equal')
plt.show()
```

```python
twitter_exec = twitter[twitter['Executive/Non-executive'] == 'Executive']
twitter_gen = twitter[twitter['Executive/Non-executive'] == 'Non-executive']
twitter_gen
```

```python
twitter_exec
```

```python
def tweets_per_day(df):
    df['Date'] = pd.to_datetime(df['Date'])
#   return df[['Tweets']].groupby(df['Date'].dt.date).count()
    tweet = df['Tweet'].groupby(df['Date'].dt.date).count() 
    d1 = pd.DataFrame(tweet)
    return d1
  # if you want output to be `Series`

df1 = tweets_per_day(twitter_gen)
df1 = df1.reset_index(level=0)
df2 = tweets_per_day(twitter_exec)
df2 = df2.reset_index(level=0)
```

```python
df1
```

```python
df1.rename(columns = {'Tweet':'Count of Tweet'}, inplace = True)
```

```python
df2
```

```python
df2.rename(columns = {'Tweet':'Count of Tweet'}, inplace = True)
```

```python
fig, ax = plt.subplots(figsize=(30,10))
sns.set(font_scale=2)
sns.lineplot(data=df1, x='Date', y='Count of Tweet', label='general',color ='b', ax=ax,legend=False)
ax2 = ax.twinx()
sns.set(font_scale=2)
sns.lineplot(data=df2, x='Date', y='Count of Tweet', label='executive',color ='g', ax=ax2,legend=False)
ax.figure.legend(fontsize=25)
plt.show()
```
